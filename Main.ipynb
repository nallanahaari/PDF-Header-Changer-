```python
import os
from PyPDF2 import PdfReader, PdfWriter
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from io import BytesIO
from decimal import Decimal

def create_new_header_page(new_header, page_width, page_height):
   
    packet = BytesIO()
    c = canvas.Canvas(packet, pagesize=(float(page_width), float(page_height)))
    c.setFont("Helvetica", 12)
    c.drawString(50, float(page_height) - 50, new_header)  # Adjust coordinates for header placement
    c.save()
    packet.seek(0)
    return PdfReader(packet).pages[0]

def process_pdf(input_file, output_file, new_header):
   
    reader = PdfReader(input_file)
    writer = PdfWriter()

    for page in reader.pages:
        
        crop_height = Decimal(50)  
        media_box = page.mediabox
        page.mediabox.upper_right = (media_box.upper_right[0], media_box.upper_right[1] - crop_height)

        
        header_page = create_new_header_page(new_header, media_box.width, media_box.height)
        page.merge_page(header_page)

        writer.add_page(page)

    
    with open(output_file, "wb") as f:
        writer.write(f)


folder_path = "sem 5 dsa lab"


new_header = "221020XX    RUDRONEEL SENGUPTA   A3"


for file_name in os.listdir(folder_path):
    if file_name.endswith(".pdf"):
        input_path = os.path.join(folder_path, file_name)
        output_path = os.path.join(folder_path, "Rudroneel_" + file_name)
        process_pdf(input_path, output_path, new_header)
        print(f"Updated headers in: {file_name}")
```
